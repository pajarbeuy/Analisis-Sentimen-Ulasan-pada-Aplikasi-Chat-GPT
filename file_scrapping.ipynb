```python
from google_play_scraper import app, reviews, Sort, reviews_all
import pandas as pd
import numpy as np

result, _ = reviews(
    'com.openai.chatgpt',
    lang='id',
    country='id',
    count=10000
)

import csv
with open('ulasan_chatgpt.csv', mode = 'w', newline='', encoding='utf-8') as file:
  writer = csv.writer(file)
  writer.writerow(['Review']) # Menulis header kolom
  for review in result:
    writer.writerow([review['content']]) #Menulis ulasan ke dalam file csv
```

```python
df_chatgpt = pd.DataFrame(result)
df_chatgpt.shape
df_chatgpt.to_csv('ulasan_aplikasi_chatgpt.csv', index=False)

#DataFrame dari scraping
df_chatgpt = pd.DataFrame(result)
jumlah_ulasan, jumlah_kolom = df_chatgpt.shape
```
